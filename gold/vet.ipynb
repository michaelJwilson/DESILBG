# Overlap

```python
hdr = '\tBXU\tG\tTMG'

print(hdr)

for c, clabel in zip([bxu, gdrops, tmg], ['BXU', 'G', 'TMG']):
    row = clabel
    
    for d, dlabel in zip([bxu, gdrops, tmg], ['BXU', 'G', 'TMG']):
        noverlap = len(np.intersect1d(c['ID'], d['ID']))

        row += '\t{}'.format(noverlap)
    
    print(row)
```
