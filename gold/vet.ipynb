```python
import pylab as pl
import numpy as np

from   astropy.table import Table
```

```python
bxu    = Table.read('/global/cscratch1/sd/mjwilson/DESILBG/GOLD/BXU/bxu.fits')
gdrops = Table.read('/global/cscratch1/sd/mjwilson/DESILBG/GOLD/G/g.fits')
tmg    = Table.read('/global/cscratch1/sd/mjwilson/DESILBG/GOLD/TMGV9/tmg_v9.fits')
```

```python
bxu.pprint()
```

# BX/U & TMG

```python
pl.plot(bxu['g'] - bxu['r'], bxu['u'] - bxu['g'], marker='.', lw=0.0, c='k', markersize=2)
pl.xlim(-5.5, 1.5)
pl.ylim(-0.5, 7.5)
pl.xlabel('g-r')
pl.ylabel('u-g')
pl.title('BX/U/NONDETECT')
```

```python
pl.plot(tmg['g'] - tmg['r'], tmg['u'] - tmg['g'], marker='.', lw=0.0, c='k', markersize=2)
pl.xlim(-5.5, 1.5)
pl.ylim(-0.5, 7.5)
pl.xlabel('g-r')
pl.ylabel('u-g')
pl.title('TMG - r priority')
```

```python
_ = pl.hist(bxu['r'], bins=np.arange(17., 25., 0.25))
pl.title('BX/U/NONDETECT')
pl.xlabel('r')
```

```python
_ = pl.hist(bxu['Z_BC03'], bins=np.arange(0.0, 6.0, 0.1))
pl.xlabel('z')
pl.title('BX/U/NONDETECT')
```

```python
_ = pl.hist(tmg['Z_BC03'], bins=np.arange(0.0, 6.0, 0.1))
pl.xlabel('z')
pl.title('TMG')
```

# g drops

```python
pl.plot(gdrops['r'] - gdrops['i'], gdrops['g'] - gdrops['r'], marker='.', lw=0.0, c='k', markersize=2)
pl.xlim(-5.5, 1.5)
pl.ylim(-0.5, 7.5)
pl.xlabel('g-r')
pl.ylabel('u-g')
pl.title('g drops')
```

```python
_ = pl.hist(gdrops['Z_BC03'], bins=np.arange(0.0, 6.0, 0.1))
pl.xlabel('z')
pl.title('g drops')
```

```python
_ = pl.hist(gdrops['i'], bins=np.arange(17., 26., 0.25))
pl.title('G/NONDETECT')
pl.xlabel('i')
```

```python
hdr = '\tBXU\tG\tTMG'

print(hdr)

for c, clabel in zip([bxu, gdrops, tmg], ['BXU', 'G', 'TMG']):
    row = clabel
    
    for d, dlabel in zip([bxu, gdrops, tmg], ['BXU', 'G', 'TMG']):
        noverlap = len(np.intersect1d(c['ID'], d['ID']))

        row += '\t{}'.format(noverlap)
    
    print(row)
```
